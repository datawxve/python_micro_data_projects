## 3. Data Cleansing, Enhancing & Preparation

The process below is where we will clean and enhance the dataset. Cleaning the dataset is making the dataset usable for modeling (e.g. missing values & mispellings). Enhancing the dataset, brings additional information to increase prediction potential or extract in from existing data (e.g. Calculations & Encoding). 


1. The rows where the column TotalCharges is missing values in 23 columns. We deal with this simply by removing them. This was justified by how small of a portion this was of the dataset. Increasing the complexity of the method to dealing with this may not reward in proportion to the cost (i.e. time & effort).


2. Change the datatype of the column 'TotalCharges' from a object into a numeric type specifically float. This was so that the column would be usable in the model. The sklearn model will not be able to use this column in the previous datatype.


3. Encode the categorical values into numerical values, making them usable for the model. This also applies the drop_first method where we use the n-1 values as we can get the same information with less data.


4. Print the columns in the datafram to verify the categroical columns were encoded properly.


5. Seperating the dataset into training, validation and testing sets. The testing set is 27% of the entire dataset.

```python
# drop ' ' value (blank but not NaN).
#
# data = data.drop(data.index[[488, 751, 752, 933, 1073,1076, 1330, 3316, 3805, 4352, 5182, 6625, 6699]])

# change the datatype of TotalCharges from object to float.
#
#data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='raise')

# encode all the categorical data.

# use the get dummies function to encode the categorical variables

# notice the drop_first attribute is used to drop one of the newly encoded columns

# having all encoded columns is unnecessary, as we can access the same information using n-1

# this was used due to being easier to implement at the moment over OneHotEncoding

# dummy features

dum_feat = ['gender', 'Partner', 'Dependents', 
            'PhoneService', 'MultipleLines', 
            'InternetService', 'OnlineSecurity', 
            'OnlineBackup', 'DeviceProtection', 
            'TechSupport', 'StreamingTV', 
            'StreamingMovies','Contract', 
            'PaperlessBilling', 'PaymentMethod', 'Churn']

# applying the dummy encoder on dummy features

data = pd.get_dummies(data, columns = dum_feat, drop_first = True)

# print the final dataset to a csv file

data.to_csv('b. model_data.csv')

# printing the columns

data.columns
```

## 5. Model Selection

Below is the model selection process. Here, six potential classifier models are compared on the dataset using a K fold cross validation method (K = 10). This was done looping through the list of models, running throught the cross validation and appending the scores to a list. These scores are then plotted on a boxplot showing mean and standard deviation. The model that performed with the most accuracy and lowest standard deviation was the Logistic Regression, although the SVM and Linear Discriminant models perfored within 2% accuracy. Based on these results, the selected model is the Logistic Regression. However, this doesn't justify invesigating the other models to be sure.

1. Model comparison loop.

```python
outcome = []
model_names = []
models = [('LogReg', LogisticRegression(solver='lbfgs',max_iter=1000)), 
          ('SVM', SVC(gamma='scale')), 
          ('DecTree', DecisionTreeClassifier()),
          ('KNN', KNeighborsClassifier()),
#          ('LinDisc', LinearDiscriminantAnalysis()),
          ('GaussianNB', GaussianNB())]

for model_name, model in models:
    k_fold_validation = model_selection.KFold(n_splits=10, random_state=7)
    results = model_selection.cross_val_score(model, X_train, y_train, cv=k_fold_validation, scoring='accuracy')
    outcome.append(results)
    model_names.append(model_name)
    output_message = "%s| Mean=%f STD=%f" % (model_name, results.mean(), results.std())
    print(output_message)
```

2. Table and plot showing the model selection results.

```python
# create a pandas DataFrame from the model selection results.

model_selection_results = pd.DataFrame(outcome).T
model_selection_results.columns = [model_names]
model_selection_results
```

3. Plot showing the performance of each model.

```python
fig = plt.figure()
fig.suptitle('Machine Learning Model Comparison')
ax = fig.add_subplot(111)
plt.boxplot(outcome)
ax.set_xticklabels(model_names)
plt.savefig('c. Log_ROC.png')
plt.show()
```

# 5. Logistic Regression & Performance Analysis

In this section we will run the Logistic Regression against the entire dataset leaving out 27% for testing. In additon, the models metric and results will be analyzed. 

1. Splitting the dataset into training and testing and checking the size of the split datasets.

2. Training the model using the 'liblinear' solver with a maximum iterations of 100.

```python
# training the model

LogReg = LogisticRegression(solver='liblinear',max_iter=100)  
LogReg.fit(X_train, y_train)
```

```python
LogReg.get_params()
```

3. Printout of the models metrics including the coefficients, the Y intercept and the R squared value.

```python
# print the coefficients, intercept and the R squared value

print(f'Coefficients: {LogReg.coef_}')
print(f'Intercept: {LogReg.intercept_}')
print(f'R^2 score: {LogReg.score(X, y)}')
```

4. Test the model by making predictions on the testing set and appending the results to a pandas dataframe.

```python
# make predictions from testing data

y_pred = LogReg.predict(X_test)

# table showing the actual vs predicted data

pred_table = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

pred_table['correct'] = pred_table['Actual'] == pred_table['Predicted']

pred_table.head(10)
```

5. Below are the models performance metrics including the Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE) and the percentage of correct predictions (Score).

```python
# error metrics model metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Percent Correct Predictions:', LogReg.score(X_test, y_test))
```

6. Confusion Matrix

```python
# print confusion matrix

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

7. Classification Report

```python
# print a classification report

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))

```

8. POC Curve

```python
# Plot the ROC Curve 

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, LogReg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, LogReg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('d. Log_ROC.png')
plt.show()
```
