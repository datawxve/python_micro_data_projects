## 4. Data Engineering

The below code processess the data and calculates new columns potentially used as features and columns in the model. Lastly, the rows with missing data were dropped. The missing data stems from the shifted columns. Currently, the newly calculated columns include:


1. Delta between the days Open and Close Price.


2. Delta between High and Low Price of the Day.


3. Stock price 1 day out.


4. Target column as a boolean of the stock going up or down 1 day out.


5. Seperating the date into 3 columns including Year, Month and Day.

```python
# remove todays data for tomorrows prediction
todays_data = data.tail(1)
data = data[:-1]

# the difference (delta) between open and close & high and low
data['delta_OpenClose'] = abs(data['Open'] - data['Close'])
data['delta_HighLow'] = abs(data['High'] - data['Low'])

todays_data['delta_OpenClose'] = abs(todays_data['Open'] - todays_data['Close'])
todays_data['delta_HighLow'] = abs(todays_data['High'] - todays_data['Low'])

# add target column (i.e. future predictions)
data['1d_target_close'] = data['Close'].shift(-1)

todays_data['1d_target_close'] = todays_data['Close'].shift(-1)


# add classification target for up or down activity
data['1d_market_mvmt'] = data['Close'] < data['1d_target_close']

todays_data['1d_market_mvmt'] = todays_data['Close'] < todays_data['1d_target_close']

# encode target
data = pd.get_dummies(data, columns = ['1d_market_mvmt'], drop_first = True)

todays_data = pd.get_dummies(todays_data, columns = ['1d_market_mvmt'], drop_first = True)

# reset the index to show the 
data = data.reset_index()

todays_data = todays_data.reset_index()

# seperating dates into seperate columns
data['year'] = pd.DatetimeIndex(data['Date']).year
data['month'] = pd.DatetimeIndex(data['Date']).month
data['day'] = pd.DatetimeIndex(data['Date']).day

todays_data['year'] = pd.DatetimeIndex(todays_data['Date']).year
todays_data['month'] = pd.DatetimeIndex(todays_data['Date']).month
todays_data['day'] = pd.DatetimeIndex(todays_data['Date']).day

#drop missing rows from shifting.
data = data.dropna()

# view the dataframe tail to verify shifting
data.tail(5)
```

## 6. Identifying Features, Target and Splitting the Dataset into Training / Testing 

```python
features = data.columns.tolist()

droplist = ['Date',
            'Open',
            'High',
            'Low',
            'delta_OpenClose',
            '1d_market_mvmt_True',
            '1d_target_close']

for x in droplist:
    features.remove(x)

# Assigning X and y variables.

# The X variable is equal to the feature dataset.

X = data[features].values
X2 = todays_data[features].values

# The y variable is equal to the target value.

y = data['1d_market_mvmt_True'].values

# split data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.27, random_state=0)
```

## 7. Model Selection

In the model selection process we will loop through a list of potnentially vaible clasification models to get an idea of the base performance of each method. The top performing models will be tested further, tweaking the parameter and features to achive better results. It is likely after further analysis there will be a marginal increase in performance identified. This will likely prove to be the model used to solve the problem in the future.

### A. Model Testing Using the K Fold Method

```python
# model selection test

outcome = []
model_names = []
models = [('LogReg', LogisticRegression(solver='lbfgs',max_iter=100000000)), 
          ('SVM', SVC(gamma='scale')), 
          #('MLP', MLPClassifier(alpha=1, max_iter=10000,hidden_layer_sizes=(100, 100), activation='relu', solver='adam')),
          ('DecTree', DecisionTreeClassifier()),
          ('KNN', KNeighborsClassifier()),
          ('LinDisc', LinearDiscriminantAnalysis()),
          ('GaussianNB', GaussianNB()),
          ('AdaB',AdaBoostClassifier(n_estimators=100, random_state=0)),
          ('SGD', linear_model.SGDClassifier(max_iter=1000, tol=1e-3))]

for model_name, model in models:
    k_fold_validation = model_selection.KFold(n_splits=10, random_state=7)
    results = model_selection.cross_val_score(model, X_train, y_train, cv=k_fold_validation, scoring='accuracy', error_score='raise')
    outcome.append(results)
    model_names.append(model_name)
    output_message = "%s| Mean=%f STD=%f" % (model_name, results.mean(), results.std())
    print(output_message)
```

### B. Model Selection  Resluts (KFold Method K = 10)

```python
# create a pandas DataFrame from the model selection results.

model_selection_results = pd.DataFrame(outcome).T
model_selection_results.columns = [model_names]
model_selection_results
```

```python
# plot model selection results

fig = plt.figure(figsize=(14, 8))
fig.suptitle('Prediction Results of K Fold Model Selection')
ax = fig.add_subplot(111)
plt.boxplot(outcome)
ax.set_xticklabels(model_names)
plt.show()
```

## 8. Model Testing Using the Split Method (Testing Size 27%)

```python
models = [('LogReg', LogisticRegression(solver='lbfgs',max_iter=100000000)), 
          ('SVM', SVC(gamma='scale')), 
          #('MLP', MLPClassifier(alpha=1, max_iter=10000,hidden_layer_sizes=(100, 100), activation='relu', solver='adam')),
          ('DecTree', DecisionTreeClassifier()),
          ('KNN', KNeighborsClassifier()),
          ('LinDisc', LinearDiscriminantAnalysis()),
          ('GaussianNB', GaussianNB()),
          ('AdaB',AdaBoostClassifier(n_estimators=100, random_state=0)),
          ('SGD', linear_model.SGDClassifier(max_iter=1000, tol=1e-3))]

PercCorr = []
model_names = []
next_day_predictions = []

for model_name, model in models:
    model_names.append(model_name)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = model.score(X_test, y_test)
    PercCorr.append(score)
    next_day_predictions.append(model.predict(X2))
    print(model_name)
    print(' ')
    print(classification_report(y_test, y_pred))
```

## 9. Predictions for Next Days Close Each Model

```python
print('see model results below:')
ziplist =  list(zip(model_names, PercCorr, next_day_predictions))
model_testing = pd.DataFrame(ziplist, columns = ['Model Name','score','next_day_predictions'])
model_testing
```
