```python
# import the necessary packages to handle urls,json files, make request and manipulate the data.

from urllib.request import urlopen
import json
import requests
import pandas as pd
```

```python
# Assigning variables for the API's and turning them into json objects

# ****** MUST CHANGE YEAR AND WEEK IN API ********

url = 'https://api.sleeper.app/v1/stats/nfl/regular/2017/3'

json_obj = urlopen(url)
data = json.load(json_obj)
```

```python
print(data)
```

```python
# change the dictionary to a pandas dataframe for easy data manipulation

new_data = pd.DataFrame(data)
new_data
```

```python
# visualize the transformed dataframe

new_data.T
```

```python
# transform the dataframe

new_data = new_data.T
```

```python
# size of the newly created tables size

new_data.shape
```

```python
# list the columns
list(new_data.columns.values)
```

```python
new_data.describe(include ='all')
```

```python
# Inserting the year and week into the dataframe

# ****** MUST CHANGE YEAR AND WEEK VARIABLES ********

new_data['year']= 2017
new_data['week']= 3
```

```python
# confirm the new columns were added

list(new_data.columns)
```

```python
# convert new dataframe to csv file for storage

# ****** MUST CHANGE FILE NAME ********

new_data.to_csv('scrapped_2017_week_3.csv', header=True)
```

```python

master_data_file = pd.read_csv('master_data_file.csv')

master_concat = pd.concat([new_data, master_data_file],sort=False)

master_concat.to_csv('master_data_file.csv', header=True)
```
