## Feature Engineering

Here we are enhancing the dataset doing the following:

1. Calculate the age of for each player in the specified season and creating a column called current_age.


2. Add cumulative sums columns for all numerical data.


3. Create a column of the mean points per game.


4. Lastly, create a  columns called high_performer to capture players who scored at or above the threshold.

```python
# calculate the current age of the player in each season
# pandas datetimeindex docs: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DatetimeIndex.html

data['birth_year'] = pd.DatetimeIndex(data['birth_date']).year

data['current_age'] = data['year'] - data['birth_year'] 
    
# add cumulative sum columns

cols = data.select_dtypes(include=np.number).columns.tolist()

for col in cols:
    txt = 'cumsum_{}'
    data[txt.format(col)] = data.sort_values('year').groupby(['full_name'])[col].cumsum()- data[col]
    
# drop the year 2009 as this is the first year in the dataset and cum sum stats are 0

data = data[data.year != 2009] 

# calculate the average points per game and add to a column

data['avg_points'] = data['pts_ppr']/data['gp']

# add target column for players scoring above threshold

pts_threshold = 10

data['high_performer'] = data['avg_points']>=pts_threshold

# hot encode target column from converting it to a numerical value

data = pd.get_dummies(data, columns = ['high_performer','position'], drop_first = True)

# create a pandas DataFrame from the model selection results.

data.dropna()
```

```python
# setting features to all numerical values

features = ['cumsum_pts_ppr','cumsum_cmp_pct', 'cumsum_fum', 'cumsum_gp', 'cumsum_gs', 'cumsum_kr', 'cumsum_kr_lng', 
            'cumsum_kr_td', 'cumsum_kr_yd', 'cumsum_kr_ypa', 'cumsum_pass_2pt', 'cumsum_pass_att', 'cumsum_pass_cmp',
            'cumsum_pass_int', 'cumsum_pass_lng', 'cumsum_pass_sack', 'cumsum_pass_sack_yds', 'cumsum_pass_td', 
            'cumsum_pass_yd', 'cumsum_pass_ypa', 'cumsum_pass_ypc', 'cumsum_pr', 'cumsum_pr_lng', 'cumsum_pr_td', 
            'cumsum_pr_yd', 'cumsum_pr_ypa', 'cumsum_rec', 'cumsum_rec_2pt', 'cumsum_rec_lng', 'cumsum_rec_pct', 
            'cumsum_rec_td', 'cumsum_rec_tgt', 'cumsum_rec_yd', 'cumsum_rec_ypr', 'cumsum_rec_ypt', 
            'cumsum_rush_2pt', 'cumsum_rush_att', 'cumsum_rush_lng', 'cumsum_rush_td', 'cumsum_rush_yd', 
            'cumsum_rush_ypa', 'cumsum_st_td', 'cumsum_off_snp', 'cumsum_st_snp', 'cumsum_pass_fd', 
            'cumsum_rush_fd', 'cumsum_td', 'cumsum_bonus_pass_yd_300', 'cumsum_bonus_pass_yd_400', 
            'cumsum_bonus_rec_rb', 'cumsum_bonus_rec_te', 'cumsum_bonus_rec_wr', 'cumsum_bonus_rec_yd_100', 
            'cumsum_bonus_rec_yd_200', 'cumsum_bonus_rush_yd_100', 'cumsum_bonus_rush_yd_200', 'cumsum_pass_inc', 
            'cumsum_pass_cmp_40p', 'cumsum_rec_40p', 'cumsum_rec_fd', 'cumsum_rush_40p', 'cumsum_bonus_pass_cmp_25', 
            'cumsum_bonus_rush_att_20', 'cumsum_bonus_rush_rec_yd_100', 'cumsum_bonus_rush_rec_yd_200', 
            'cumsum_fgm_yds_over_30', 'cumsum_pass_td_40p', 'cumsum_rec_0_4', 'cumsum_rec_10_19', 
            'cumsum_rec_20_29', 'cumsum_rec_30_39', 'cumsum_rec_5_9', 'cumsum_rec_td_40p', 'cumsum_rush_td_40p', 
            'position_RB', 'position_TE', 'position_WR']
```

## Model Selection

Below various classification models are tested against our dataset. This was done by running the models in a K Fold vaildation with a K value of 8. The testing results are appended to a table for future analysis. The models Tested includeds:


1. Logistic Regression


2. Support Vector Classifier


3. Multi Layer Perceptron Classifier


4. Decistion Tree Classifier


5. Linear Discrimination Analysis


6. Gausian NB


7. AdaBoost Classifier


8. Stochastic Gradient Descent

```python
# model selection test

outcome = []
model_names = []
models = [('LogReg', LogisticRegression(solver='lbfgs',max_iter=100000000)), 
          ('SVM', SVC(gamma='scale')), 
          ('MLP', MLPClassifier(alpha=1, max_iter=1000,hidden_layer_sizes=(100, 100, 100), activation='relu', solver='adam')),
          ('DecTree', DecisionTreeClassifier()),
          ('KNN', KNeighborsClassifier()),
          ('LinDisc', LinearDiscriminantAnalysis()),
          ('GaussianNB', GaussianNB()),
          ('AdaB',AdaBoostClassifier(n_estimators=100, random_state=0)),
          ('SGD', linear_model.SGDClassifier(max_iter=1000, tol=1e-3))]

for model_name, model in models:
    k_fold_validation = model_selection.KFold(n_splits=8, random_state=7)
    results = model_selection.cross_val_score(model, X_train, y_train, cv=k_fold_validation, scoring='accuracy', error_score='raise')
    outcome.append(results)
    model_names.append(model_name)
    output_message = "%s| Mean=%f STD=%f" % (model_name, results.mean(), results.std())
    print(output_message)
```

```python
# create a pandas DataFrame from the model selection results.

model_selection_results = pd.DataFrame(outcome).T
model_selection_results.columns = [model_names]
model_selection_results
```

```python
# plot model selection results
fig = plt.figure(figsize=(14, 8))
fig.suptitle('Machine Learning Model Comparison')
ax = fig.add_subplot(111)
plt.boxplot(outcome)
ax.set_xticklabels(model_names)
plt.savefig('c. Log_ROC.png')
plt.show()
```
